```python
import requests
import json
import pandas as pd
from datetime import date
from datetime import timedelta
```

```python
id = 420100
type = 'move_out'
```

```python
url = 'http://huiyan.baidu.com/migration/historycurve.jsonp?dt=city&id={id}&type={type}'.format(id = id,type=type)
```

```python
#get response as string
file = requests.get(url)
```

```python
#read data from file 
data = json.loads(file.text.split('(')[1][:-1])
```

```python
dict = data['data']['list']
```

```python
dict
```

```python
col = str(id) + '_' + type + '_index'
```

```python
df = pd.DataFrame(dict,index=[0])
```

```python
day_ = int(str(date.today() - date(2020,1,1)).split(' ')[0])
df_2019 = df.iloc[:,:-day_]
df_2020 = df.iloc[:,df.shape[1] - day_:]
```
